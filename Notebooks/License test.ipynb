### Used to create and validate license files

```python
license={
        'user ID':'',
        'expiry date':'',
        'computer ID':'',
        'license reg path':'',
        'license reg ID':'',
        'signature':''
        }
```

```python
license['computer ID'] = str(hex(uuid.getnode()))
license['user ID'] = 'KullmannLab'
license['expiry date'] = '9999-12-31'
```

```python
fname = '/home/mfpleite/Documents/PyEcogLicense.txt'
fname = 'C:\\Users\\mfple\\PycharmProjects\\dist\\PyEcogLicense.txt'
with open(fname,'r') as f:
    license = json.load(f)
```

```python
keys =  ['user ID', 'expiry date', 'computer ID', 'license reg path', 'license reg ID']
licensestring = ''.join([license[k] for k in keys]).encode('utf-8')
```

```python
signature = b64encode(rsa.sign(licensestring, privkey, 'SHA-1')).decode('ascii')
license['signature'] = signature
print(signature)
```

```python
with open(fname,'w') as f:
    json.dump(license,f)
```
